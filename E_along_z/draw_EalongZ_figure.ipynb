```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

plt.rc('font', size=10)
tick_size = 12
plt.rc('xtick', labelsize=tick_size) 
plt.rc('ytick', labelsize=tick_size) 

file_1 = 'EalongZ_data1.txt'
file_2 = 'EalongZ_data2.txt'

# Load file as array: data
data_1 = np.loadtxt(file_1, delimiter=',', skiprows = 3)
data_2 = np.loadtxt(file_2, delimiter=',', skiprows = 3)

# Print datatype of digits
print(data_1[:,0])
print(data_2)
print(len(data_1))
```
