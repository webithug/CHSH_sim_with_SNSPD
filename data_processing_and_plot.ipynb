```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
plt.rc('font', size=10)
tick_size = 12
plt.rc('xtick', labelsize=tick_size) 
plt.rc('ytick', labelsize=tick_size) 


file_1 = 'NbN_1.txt'
file_2 = 'NbN_2.txt'

# Load file as array: digits
data_1 = np.loadtxt(file_1, delimiter=',', skiprows = 1)
data_2 = np.loadtxt(file_2, delimiter=',', skiprows = 1)

# Print datatype of digits
# print(data_1[:,0])
# print(data_2)

```

```python
# interpolate data_1
x_1 = data_1[:,0]
y_1 = data_1[:,1]

cs_1 = CubicSpline(x_1, y_1) # a function
xs_1 = np.arange(0, 360, 0.001)

fig, ax_1 = plt.subplots(figsize=(6.5, 4))
ax_1.plot(x_1, y_1, 'o', label='data')
ax_1.plot(xs_1, cs_1(xs_1), label="S")

print(cs_1(21.5448 ))
```

```python
# interpolate data_2
x_2 = data_2[:,0]
y_2 = data_2[:,1]

cs_2 = CubicSpline(x_2, y_2) # a function
xs_2 = np.arange(0, 360, 0.001)

line_width = 8
plt.plot(xs_2, cs_2(xs_2), label="Vertical", linewidth = line_width)
plt.plot(xs_1, cs_1(xs_1), label="Horizontal", linewidth =line_width)
plt.plot(xs_1, cs_1(xs_1)+cs_2(xs_2), label="Sum", linewidth =line_width)
# plt.legend(loc='upper left')
# plt.xlabel("polarization angle")
# plt.ylabel("absorption")
plt.ylim(0,1)
plt.xlim(0,360)


print(cs_2(21.5448))
```
